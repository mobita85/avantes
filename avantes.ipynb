```python
filename = "190107 - spectra - 10 - 100 ms - B"
file_ = path+filename+ ".xlsx"
data = open(file_)
```

```python
#import required libraries
from openpyxl import load_workbook
#read  from excel file
wb = load_workbook(file_)
sheet_1 = wb.get_sheet_by_name('0410019S1_m')
q = sheet_1.max_row
p = sheet_1.max_column
data  = empty([q, p])
print data.shape
```

```python
for i in range(0,sheet_1.max_column):
    for j in range(7,sheet_1.max_row):
        data[j,i] = sheet_1.cell(row=j, column=i+1).value
```

```python
'''Save data'''
save(path+filename, data)
```

```python
data = load(path+filename+".npy")
```

```python
fig, ax = plt.subplots()
on = 5
ax.plot(data[:,0],data[:,on],'c-', label='0 ms')
#ax.plot(data[:,0],data[:,8],'r-', label='100 ms')
ax.plot(data[:,0],data[:,on+1],'m-', label='100 ms')
ax.plot(data[:,0],data[:,on+2],'k--', label='200 ms')
ax.plot(data[:,0],data[:,on+3],'b-', label='300 ms')
ax.plot(data[:,0],data[:,on+4],'g-', label='400 ms')
ax.plot(data[:,0],data[:,on+5],'r-', label='500 ms')
ax.plot(data[:,0],data[:,on+6],'y-', label='600 ms')
ax.plot(data[:,0],data[:,on+7],'r--', label='700 ms')
ax.plot(data[:,0],data[:,on+8],'b-', label='800 ms')
ax.plot(data[:,0],data[:,on+9],'m--', label='900 ms')
ax.plot(data[:,0],data[:,on+10],'k-', label='1000 ms')
#ax.plot(data[:,0],data[:,12],'r-', label='500 ms')
#ax.plot(data[:,0],data[:,12],'r--')
#ax.plot(data[:,0],data[:,13],'r--')
ax.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.xlim(500,1000)
plt.savefig(path+filename+' - 0-1s'+'.png', bbox_inches='tight')
plt.savefig(path+filename+' - 0-1s'+'.pdf', bbox_inches='tight')
plt.show()
```
